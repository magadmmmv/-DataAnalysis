**–£—Ä–æ–∫ 1. –ê–ª–≥–æ—Ä–∏—Ç–º –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –ì—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫**

–ó–∞–¥–∞—á–∞: –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –±–∞–ª–ª–æ–≤ –ï–ì–≠ —É—á–µ–Ω–∏–∫–∞ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –∫–æ–ª-–≤–∞ –ª–µ—Ç —Å—Ç–∞–∂–∞ –µ–≥–æ —Ä–µ–ø–µ—Ç–∏—Ç–æ—Ä–∞

```python
import numpy as np
```

```python
X = np.array([[ 1,  1],
              [ 1,  1],
              [ 1,  2],
              [ 1,  5],
              [ 1,  3],
              [ 1,  0],
              [ 1,  5],
              [ 1, 10],
              [ 1,  1],
              [ 1,  2]])
```

```python
X.shape
```

```python
y = [45, 55, 50, 55, 60, 35, 75, 80, 50, 60]
```

```python
def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err
```

1. –ü–æ–¥–æ–±—Ä–∞—Ç—å —Å–∫–æ—Ä–æ—Å—Ç—å –æ–±—É—á–µ–Ω–∏—è (eta) –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π

```python
n = X.shape[1]

# –ò–∑–º–µ–Ω–µ–Ω–∏—è
n_iter = 100
c = 0.1
eta = c
# –ò–∑–º–µ–Ω–µ–Ω–∏—è

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
    for k in range(W.shape[0]):
        W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    if i % 10 == 0:
        # –ò–∑–º–µ–Ω–µ–Ω–∏—è (–º–µ–Ω—è–µ–º —à–∞–≥ –Ω–∞ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∑–∞–¥–∞–Ω–Ω–æ–π –∫–æ–Ω—Å—Ç–∞–Ω—Ç—ã –Ω–∞ –∏—Ç–µ—Ä–∞—Ü–∏—é)
        eta = c/(i+1)
        # –ò–∑–º–µ–Ω–µ–Ω–∏—è
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 2)}')
```

2*. –í —ç—Ç–æ–º –∫–æ–¥–µ –º—ã –∏–∑–±–∞–≤–ª—è–µ–º—Å—è –æ—Ç –∏—Ç–µ—Ä–∞—Ü–∏–π –ø–æ –≤–µ—Å–∞–º, –Ω–æ —Ç—É—Ç –µ—Å—Ç—å –æ—à–∏–±–∫–∞, –∏—Å–ø—Ä–∞–≤—å—Ç–µ –µ–µ

```python
n = X.shape[1]

eta = 1e-2 
n_iter = 100

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
#     for k in range(W.shape[0]):
#         W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    # –ò–ó–ú–ï–ù–ï–ù–ò–Ø
    W[0] -= eta * (1/n * 2 * np.dot(X[:, 0], y_pred - y))
    W[1] -= eta * (1/n * 2 * np.dot(X[:, 1], y_pred - y))
    # –ò–ó–ú–ï–ù–ï–ù–ò–Ø
    if i % 10 == 0:
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err,2)}')
```

3*. –í–º–µ—Å—Ç–æ —Ç–æ–≥–æ, —á—Ç–æ–±—ã –∑–∞–¥–∞–≤–∞—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π, –∑–∞–¥–∞–π—Ç–µ –¥—Ä—É–≥–æ–µ —É—Å–ª–æ–≤–∏–µ –æ—Å—Ç–∞–Ω–æ–≤–∞ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ - –∫–æ–≥–¥–∞ –≤–µ—Å–∞ –ø–µ—Ä–µ—Å—Ç–∞—é—Ç –∏–∑–º–µ–Ω—è—Ç—å—Å—è –º–µ–Ω—å—à–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–≥–æ –ø–æ—Ä–æ–≥–∞  ùúñ .

```python
n = X.shape[1]

eta = 1e-2
n_iter = 100

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

# –ò–ó–ú–ï–ù–ï–ù–ò–Ø-------------------------------
# –∫—Ä–∏—Ç–µ—Ä–∏–π —Å—Ö–æ–¥–∏–º–æ—Å—Ç–∏ (—Ä–∞–∑–Ω–∏—Ü–∞ –≤–µ—Å–æ–≤, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–π –∞–ª–≥–æ—Ä–∏—Ç–º –æ—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ—Ç—Å—è)
min_weight_dist = 1e-4

# –∑–∞–¥–∞–¥–∏–º –Ω–∞—á–∞–ª—å–Ω—É—é —Ä–∞–∑–Ω–∏—Ü—É –≤–µ—Å–æ–≤ –±–æ–ª—å—à–∏–º —á–∏—Å–ª–æ–º
weight_dist = np.inf

# —Å—á–µ—Ç—á–∏–∫ –∏—Ç–µ—Ä–∞—Ü–∏–π
iter_num = 0

while weight_dist > min_weight_dist:
# –ò–ó–ú–ï–ù–ï–ù–ò–Ø-------------------------------------    

# for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
#     for k in range(W.shape[0]):
#         W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))

    # –ò–ó–ú–ï–ù–ï–ù–ò–Ø-------------------------------------
    new_w = W - 2 * eta * np.dot(X.T, (np.dot(X, W) - y)) / n
#     W[0] -= eta * (1/n * 2 * np.dot(X[:, 0], y_pred - y))
#     W[1] -= eta * (1/n * 2 * np.dot(X[:, 1], y_pred - y))

    weight_dist = np.linalg.norm(new_w - W, ord=2)
    
    iter_num += 1
    W = new_w
    # –ò–ó–ú–ï–ù–ï–ù–ò–Ø-------------------------------------
    
#     if i % 10 == 0:
    print(f'Iteration #{iter_num}: W_new = {W}, MSE = {round(err,2)}')
```
